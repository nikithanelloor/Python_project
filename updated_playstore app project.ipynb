```python
## Q.12
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
## initialize the linear regression model
model= LinearRegression()
```

```python
## Train the model
model.fit(x_train,y_train)
```

```python
## Converting string to float :'Free'
x_train['Price']=x_train['Price'].replace('Free','0').astype(float)
x_test['Price']=x_test['Price'].replace('Free','0').astype(float)
```

```python
## Converting 'Type' column into numeric
x_train['Type']=x_train['Type'].replace({'Free':0,'Paid':1})
x_test['Type']=x_test['Type'].replace({'Free':0,'Paid':1})
```

```python
print(x_train['Type'].unique())
```

```python
## Fit the model on Training data
model.fit(x_train,y_train)
```

```python
## Predict on the training and testing sets
y_train_pred= model.predict(x_train)
y_test_pred= model.predict(x_test)
```

```python
## Evalute the model performance
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
```

```python
print(f"Training Mean Squared Error:{train_mse:.2f}")
print(f"Testing Mean Squared Error:{test_mse:.2f}")
print(f"Training R-Squared:{train_r2:.2f}")
print(f"Tesing R-Squared:{test_r2:.2f}")
```

```python
plt.figure(figsize=(12,6))
## Training Data
plt.subplot(1,2,1)
plt.scatter(y_train,y_train_pred,alpha=0.5)
plt.plot([min(y_train),max(y_train)],[min(y_train),max(y_train)],color='red')
plt.title('Training Data: Actual vs Predicted Rating')
plt.xlabel('Actual Rating')
plt.ylabel('Predicted Rating')
## Testing Data
plt.subplot(1,2,2)
plt.scatter(y_test,y_test_pred,color='green',alpha=0.5)
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],color='red')
plt.title('Test Data: Actual vs Predicted Rating')
plt.xlabel('Actual Rating')
plt.ylabel('Predicted Rating')

plt.show()
```
